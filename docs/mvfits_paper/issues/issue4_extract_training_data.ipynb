# Extraction of Training Data for Spherical-Spheroidal mmixing coefficients

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
# The Important Things
from positive import *
from matplotlib.pyplot import *
from numpy import *
import positive
import pickle
```

## Load fit object

```python
fname = positive.__path__[0]+'/data/ysmodels.pickle'
with open( fname , 'rb') as fit_file:
     fit_object_dict = pickle.load( fit_file )
```

## Extract and plot training data for each mode

```python
for k in fit_object_dict:
    
    #
    fo = fit_object_dict[k]
    
    #
    domain_data = fo.domain
    range_data = fo.scalar_range
    
    #
    figure( figsize=2*figaspect(0.618/2) )
    
    subplot(1,2,1)
    plot( domain_data, range_data.real, '-o', label='training data' )
    xlabel(r'$\kappa$')
    ylabel(r'Re $\sigma_{\bar{\ell}\bar{m}\ell m n}$')
    title(k)
    legend()
    
    subplot(1,2,2)
    plot( domain_data, range_data.imag, '-o', label='training data' )
    xlabel(r'$\kappa$')
    ylabel(r'Im $\sigma_{\bar{\ell}\bar{m}\ell m n}$')
    title(k)
    legend()
```
