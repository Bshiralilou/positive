#### Setup Environment

```python
# Setup ipython environment
%load_ext autoreload
%autoreload 2
%matplotlib inline
# Setup plotting backend
import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 0.8
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 12
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['axes.titlesize'] = 20
from matplotlib.pyplot import *
# Import needed packages from positive
from positive.plotting import sYlm_mollweide_plot
from positive import alert
# Import other usefuls
from os.path import expanduser
from numpy import array
```

#### Plot the Real Part of a Single Spherical Harmonic

```python

# Define l,m
l,m = 3,2

# Use plotting function from positive
# NOTE that the spin weight can be input using a keyword
sYlm_mollweide_plot(l,m,N=100,form='real')
```

#### Plot a single spherical harmonic in many forms

```python
# Define l,m
l,m = 3,2

# List forms that sYlm_mollweide_plot understands
forms = ['real','imag','abs','+','x']

# Setup Axes
fig,axarr = subplots( 1, len(forms), subplot_kw=dict(projection='mollweide'), figsize = 8*array([len(forms),1]) )

# Plot each form
for form,ax in zip(forms,axarr):
    sYlm_mollweide_plot( l,m,N=100,form=form,ax=ax,colorbar_shrink=1.0 )
    
#
savefig( expanduser('~/Desktop/sYlm_mollweide_plot_many_forms_l%im%i.pdf'), pad_inches=0, bbox_inches='tight' )
```

#### Plot many forms for many sets of l,m

```python

# Define list of l,m of interest
lmlist = [ (2,2), (2,1), (2,0), (3,3), (3,2), (4,4), (4,3), (5,5), (5,4) ]

# List forms that sYlm_mollweide_plot understands
forms = ['real','imag','abs','+','x']

# Setup Axes
fig,axarr = subplots( len(lmlist), len(forms), subplot_kw=dict(projection='mollweide'), figsize = 8*array([len(forms),1.1*len(lmlist)]) )

# Plot each form
for j,(l,m) in enumerate(lmlist):
    alert('(l,m) = (%i,%i)'%(l,m),'plot_many',header=True)
    for k,form in enumerate(forms):
        print form,
        ax = axarr[j,k]
        sYlm_mollweide_plot( l,m,N=100,form=form,ax=ax,colorbar_shrink=1.0 )
    print ''
#
savefig( expanduser('~/Desktop/sYlm_mollweide_batch.pdf'), pad_inches=0, bbox_inches='tight' )
```
