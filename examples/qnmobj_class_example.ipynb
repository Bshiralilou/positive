```python
#
M = 1.0
a = 0.7
l = 3
m = -1
n = 0
```

## Test p and m without NR convention

```python

#
foo0 = qnmobj(M,a,l, m,n,verbose=True,use_nr_convention=False)
bar0 = qnmobj(M,a,l,-m,n,verbose=True,use_nr_convention=False)

#
ax = foo0.plot_slm()
bar0.plot_slm(ax=ax,ls='--');
```

## Test p and m with NR convention

```python
# For m

#
foo1 = qnmobj(M,a,l,m,n,p= 1,verbose=True,use_nr_convention=True)
bar1 = qnmobj(M,a,l,m,n,p=-1,verbose=True,use_nr_convention=True)

#
ax = foo1.plot_slm()
bar1.plot_slm(ax=ax,ls='--');
```

```python
# Now for -m

#
foo2 = qnmobj(M,a,l,-m,n,p=1,verbose=True,use_nr_convention=True)
bar2 = qnmobj(M,a,l,-m,n,p=-1,verbose=True,use_nr_convention=True)

#
ax = foo2.plot_slm()
bar2.plot_slm(ax=ax,ls='--');
```

## Tabulate and compare QNM frequencies

```python
#
print '\tConvention',
print '\t\t\t (l,m,n,p)',
print '\t\t\tQNM frequency'
print '--'*50
for foo in [foo0,bar0,foo1,foo2,bar1,bar2]:
    for mm in [m,-m]:
        if foo.m==mm:
            print'Use NR Convention = ',foo.__use_nr_convention__,'\t \t (%i,%i,%i,%i)  \t=  '%(foo.l,foo.m,foo.n,foo.p),foo.cw
```
