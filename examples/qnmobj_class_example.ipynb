## Test p and m without NR convention

```python

#
foo0 = qnmobj(M,a,l, m,n,verbose=True,use_nr_convention=False)
bar0 = qnmobj(M,a,l,-m,n,verbose=True,use_nr_convention=False)

#
ax = foo0.plot_slm()
bar0.plot_slm(ax=ax,ls='--');
```

## Test p and m with NR convention

```python
# For m

#
foo1 = qnmobj(M,a,l,m,n,p= 1,verbose=True,use_nr_convention=True)
bar1 = qnmobj(M,a,l,m,n,p=-1,verbose=True,use_nr_convention=True)

#
ax = foo1.plot_slm()
bar1.plot_slm(ax=ax,ls='--');
```

```python
# Now for -m

#
foo2 = qnmobj(M,a,l,-m,n,p=1,verbose=True,use_nr_convention=True)
bar2 = qnmobj(M,a,l,-m,n,p=-1,verbose=True,use_nr_convention=True)

#
ax = foo2.plot_slm()
bar2.plot_slm(ax=ax,ls='--');
```

## Tabulate and compare QNM frequencies

```python
#
print '\tConvention',
print '\t\t\t (l,m,n,p)',
print '\t\t\tQNM frequency'
print '--'*50
for foo in [foo0,bar0,foo1,foo2,bar1,bar2]:
    for mm in [m,-m]:
        if foo.m==mm:
            print'Use NR Convention = ',foo.__use_nr_convention__,'\t \t (%i,%i,%i,%i)  \t=  '%(foo.l,foo.m,foo.n,foo.p),foo.cw
```

## Calculate the adjoint spheroidal harmonic for the specified np subset
Spheroidal harminics allow the construction of bi-orthogonal sets along fixed values of p and n

```python
#
a = 0.7
n = 0
p = 1
s = -2
m = 2
lmin = 2 # max(abs(s),abs(m))
lmax = lmin+6
lrange = range( lmin , lmax+1 )
```

```python
#
foo = calc_adjoint_slm_subset(a,m,n,p,s,lrange,full_output=True)
```

```python
print foo.keys()
```

```python
theta = foo['theta']
S     = foo[        'spheroidal_of_theta_dict']
aS    = foo['adjoint_spheroidal_of_theta_dict']
```

```python

#
nrow = len(S)
ncol = 2

#
fig,ax = subplots( nrow, ncol, figsize=3*figaspect(0.618 * nrow / 2), sharex=True )
ax = ax.flatten()

#
k = 0
for l in lrange:
    
    #
    sca(ax[k])
    plot( theta, abs( S[l,m,n,p]),label=r'$S$',color='k',lw=1,ls='--' )
    plot( theta, abs(aS[l,m,n,p]),label=r'$\tilde{S}$',color='dodgerblue',lw=1,ls='-' )
    xlim(lim(theta))
    if l==lrange[-1]: xlabel(r'$\theta$')
    ylabel(r'$|S_{%i m n p}|$'%l)
    legend()
    k+= 1
    
    #
    sca(ax[k])
    plot( theta, unwrap(angle( S[l,m,n,p])),label=r'$S$',color='k',lw=1,ls='--' )
    plot( theta, unwrap(angle(aS[l,m,n,p])),label=r'$\tilde{S}$',color='dodgerblue',lw=1,ls='-' )
    if l==lrange[-1]: xlabel(r'$\theta$')
    ax[k].yaxis.tick_right()
    ylabel(r'$\arg(S_{%i m n p})$'%l)
    legend()
    k+= 1
    
```
