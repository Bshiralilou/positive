# Testing for rlm_sequence function
for the calculation of homogeneous solutions to Teukolsky's radial equation

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
# The Important Things
from matplotlib.pyplot import *
from numpy import *
from positive import *
```

```python
#
M = 1.0
a = 0.998
l = 2
m = 2
n = 0
```

```python
# Create QNM object
foo = qnmobj(M,a,l, m,n,verbose=True,use_nr_convention=not True,calc_rlm=False)
```

```python

#
cw = foo.cw
sc = foo.sc

#
span = 90

#
b1 = rlm_sequence_forwards(a/2,cw*2,sc,l,m,s=-2,span=span)
b2 = rlm_sequence_backwards(a/2,cw*2,sc,l,m,s=-2,span=span)
```

```python

figure( figsize=2*figaspect(0.618) )

lvals = sort(b1.keys())
b1vals = array([ b1[k] for k in lvals ])
b2vals = array([ b2[k] for k in lvals ])

plot( lvals, abs(b1vals), ls='-', marker='o',mfc='none',ms=10 )
plot( lvals, abs(b2vals), ls='-', marker='o', ms=6 )

yscale('log')
```
